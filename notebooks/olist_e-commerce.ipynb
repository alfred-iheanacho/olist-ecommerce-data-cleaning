## Missing Value Assessment

Missing values can reduce data quality and negatively impact future analysis.

This step identifies:

- Columns containing missing values
- Percentage of missing records
- Potential strategies for handling incomplete data

Appropriate actions will be taken depending on business relevance and data availability.

```python
# =====================================================
# STEP 3.1 – MISSING VALUE ANALYSIS
# =====================================================

print("MISSING VALUE ANALYSIS\n")

for name, df in datasets.items():
    print("=" * 70)
    print(f"{name} Dataset")
    print("=" * 70)

    missing_count = df.isnull().sum()
    missing_percent = (df.isnull().sum() / len(df)) * 100

    missing_table = pd.DataFrame({
        "Missing Count": missing_count,
        "Missing %": missing_percent.round(2)
    })

    missing_table = missing_table[missing_table["Missing Count"] > 0]

    if len(missing_table) == 0:
        print("No missing values.\n")
    else:
        display(missing_table)

    print("\n")
```

## Duplicate Record Detection

Duplicate records can distort analytical results and create misleading business insights.

This step identifies duplicated observations across the datasets and removes unnecessary duplicates while preserving valid business records.

```python
# ===============================================
# 3.2 ROW DUPLICATES VALIDATION
# ===============================================

print("DATA DUPLICATION VALIDATION\n")

# Row duplicate check
datasets = {
    "Customers": customers,
    "Orders": orders,
    "Order Items": order_items,
    "Payments": payments,
    "Products": products,
    "Sellers": sellers,
    "Reviews": reviews,
    "Category Translation": category_translation
}

print("----- ROW DUPLICATE CHECK -----\n")

for name, df in datasets.items():
    duplicates = df.duplicated().sum()
    print(f"{name} - Duplicate Rows: {duplicates}")

print("\n")


# ===============================================
# 3.2.1 PRIMARY / COMPOSITE KEY DUPLICATE CHECK
# ===============================================

print("----- KEY DUPLICATE CHECK -----\n")

# Customers
print("Customers - customer_id duplicates:",
      customers['customer_id'].duplicated().sum())

# Orders
print("Orders - order_id duplicates:",
      orders['order_id'].duplicated().sum())

# Order Items (composite key)
print("Order Items - (order_id + order_item_id) duplicates:",
      order_items.duplicated(subset=['order_id','order_item_id']).sum())

# Payments (composite key)
print("Payments - (order_id + payment_sequential) duplicates:",
      payments.duplicated(subset=['order_id','payment_sequential']).sum())

# Products
print("Products - product_id duplicates:",
      products['product_id'].duplicated().sum())

# Sellers
print("Sellers - seller_id duplicates:",
      sellers['seller_id'].duplicated().sum())

# Reviews (composite key)
print("Reviews - (review_id + order_id) duplicates:",
      reviews.duplicated(subset=['review_id','order_id']).sum())

# Category translation
print("Category Translation - product_category_name duplicates:",
      category_translation['product_category_name'].duplicated().sum())
```

```python
# =====================================================
# Task 4.1 – DATE CONVERSION
# =====================================================

# Orders date columns
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

# Order Items
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])

# Reviews
reviews['review_creation_date'] = pd.to_datetime(reviews['review_creation_date'])
reviews['review_answer_timestamp'] = pd.to_datetime(reviews['review_answer_timestamp'])

print("All date columns successfully converted.")

print(orders.dtypes)
print(order_items.dtypes)
print(reviews.dtypes)
```

```python
# 4.1.1 Date Null Check

for dataset_name, cols in date_columns.items():
    df = datasets[dataset_name]
    print(f"\n{dataset_name} Date Null Check")

    for col in cols:
        print(f"{col} NaT count:", df[col].isna().sum())
```

```python
# 4.2 Imputation using logically preceding timestamps

orders["order_approved_at"] = orders["order_approved_at"].fillna(
    orders["order_purchase_timestamp"]
)

orders["order_delivered_carrier_date"] = orders["order_delivered_carrier_date"].fillna(
    orders["order_approved_at"]
)

orders["order_delivered_customer_date"] = orders["order_delivered_customer_date"].fillna(
    orders["order_estimated_delivery_date"]
)
```

```python
# 4.2.1 Date Null Check

for dataset_name, cols in date_columns.items():
    df = datasets[dataset_name]
    print(f"\n{dataset_name} Date Null Check")

    for col in cols:
        print(f"{col} NaT count:", df[col].isna().sum())
```

```python
# 4.2.2 Imputation using descriptive placeholders.

reviews["review_comment_title"] = reviews["review_comment_title"].fillna("No title provided")

reviews["review_comment_message"] = reviews["review_comment_message"].fillna("No comment provided")
```

```python
# 4.3 Category Cleaning & Translation Consistency

print("=" * 60)
print("Category Cleaning & Translation Update")
print("=" * 60)

# Step 1: Replace missing categories
products["product_category_name"] = (
    products["product_category_name"]
    .fillna("unknown")
)

# Step 2: Add missing categories to translation table
missing_categories = pd.DataFrame({
    "product_category_name": [
        "portateis_cozinha_e_preparadores_de_alimentos",
        "pc_gamer",
        "unknown"
    ],
    "product_category_name_english": [
        "portable_kitchen_food_processors",
        "pc_gamer",
        "unknown"
    ]
})

category_translation = pd.concat(
    [category_translation, missing_categories],
    ignore_index=True
)

# Step 3: Quick validation
print("Missing values in products:", products["product_category_name"].isna().sum())
print("Total categories in translation table:", category_translation.shape[0])
```

```python
# 4.4 Handle Missing Product Attributes (Numeric)

print("=" * 60)
print("Handling Missing Product Attributes")
print("=" * 60)

numeric_columns = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

# Step 1: Fill missing values with median
products[numeric_columns] = products[numeric_columns].fillna(
    products[numeric_columns].median()
)

# Step 2: Validation check
print("\nRemaining missing values:")
print(products[numeric_columns].isnull().sum())
```

```python
# 4.5 Case & Space Check + Text Standardisation

print("=" * 60)
print("Text Quality Check & Standardisation")
print("=" * 60)

def check_and_clean(df, column):
    print(f"\nChecking {column}")

    # Check issues BEFORE cleaning
    stripped = df[column].str.strip()
    space_issues = (df[column] != stripped).sum()
    unique_before = df[column].nunique()

    print("Values with leading/trailing spaces:", space_issues)
    print("Unique values BEFORE cleaning:", unique_before)

    # Cleaning step (strip + lowercase)
    df[column] = df[column].str.strip().str.lower()

    # Check AFTER cleaning
    unique_after = df[column].nunique()
    print("Unique values AFTER cleaning:", unique_after)

# Apply to datasets
check_and_clean(customers, "customer_city")
check_and_clean(sellers, "seller_city")
check_and_clean(products, "product_category_name")
```

```python
# =====================================================
# 4.6 – PRIMARY KEY AND FOREIGN KEY REVALIDATION
# =====================================================

print("PRIMARY KEY VALIDATION\n")

print("Customers - customer_id unique:",
      customers['customer_id'].nunique() == len(customers))

print("Orders - order_id unique:",
      orders['order_id'].nunique() == len(orders))

print("Order Items - composite PK (order_id + order_item_id) unique:",
      order_items[['order_id','order_item_id']].drop_duplicates().shape[0] == len(order_items))

print("Payments - composite PK (order_id + payment_sequential) unique:",
      payments[['order_id','payment_sequential']].drop_duplicates().shape[0] == len(payments))

print("Products - product_id unique:",
      products['product_id'].nunique() == len(products))

print("Sellers - seller_id unique:",
      sellers['seller_id'].nunique() == len(sellers))

print("Reviews - composite PK (review_id + order_id) unique:",
      reviews[['review_id','order_id']].drop_duplicates().shape[0] == len(reviews))

print("Category Translation - product_category_name unique:",
      category_translation['product_category_name'].nunique() == len(category_translation))


# =====================================================
# FOREIGN KEY VALIDATION
# =====================================================

print("\nFOREIGN KEY VALIDATION\n")

print("Orders.customer_id exists in Customers:",
      orders['customer_id'].isin(customers['customer_id']).all())

print("Order Items.order_id exists in Orders:",
      order_items['order_id'].isin(orders['order_id']).all())

print("Order Items.product_id exists in Products:",
      order_items['product_id'].isin(products['product_id']).all())

print("Order Items.seller_id exists in Sellers:",
      order_items['seller_id'].isin(sellers['seller_id']).all())

print("Payments.order_id exists in Orders:",
      payments['order_id'].isin(orders['order_id']).all())

print("Reviews.order_id exists in Orders:",
      reviews['order_id'].isin(orders['order_id']).all())

print("Products.product_category_name exists in Category Translation:",
      products['product_category_name'].isin(category_translation['product_category_name']).all())


# =====================================================
# RELATIONSHIP VALIDATION (1:M CHECKS)
# =====================================================

print("\nRELATIONSHIP VALIDATION\n")

print("Customers → Orders (1:M):",
      orders['customer_id'].nunique() <= len(orders))

print("Orders → Order Items (1:M):",
      order_items['order_id'].nunique() <= len(order_items))

print("Orders → Payments (1:M):",
      payments['order_id'].nunique() <= len(payments))

print("Orders → Reviews (1:M) using composite PK:",
      reviews[['review_id','order_id']].drop_duplicates().shape[0] == len(reviews)
      and reviews['order_id'].nunique() <= len(reviews))
```

```python
# =====================================================
# 5.1 – LATE DELIVERY FLAG
# =====================================================

# Calculate delivery delay in days
orders['delivery_delay_days'] = (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.days

# A late delivery occurs if actual delivery > estimated delivery
orders['late_delivery'] = orders['delivery_delay_days'].apply(lambda x: 1 if x > 0 else 0)

print("Late delivery counts:")
orders['late_delivery'].value_counts()
```

```python
# 5.2 – TOTAL ORDER VALUE

order_items['total_item_value'] = order_items['price'] + order_items['freight_value']

order_revenue = (
    order_items
    .groupby('order_id')['total_item_value']
    .sum()
    .reset_index()
)

order_revenue.rename(columns={'total_item_value': 'order_total_value'}, inplace=True)

# Prevent duplicate column issue
orders = orders.drop(columns=['order_total_value'], errors='ignore')

orders = orders.merge(order_revenue, on='order_id', how='left')

print("Sample Order Revenue:")
print(orders[['order_id','order_total_value']].head())
```

```python
# =====================================================
# 5.3 – CUSTOMER LIFETIME VALUE
# =====================================================

customer_revenue = (
    orders
    .groupby('customer_id')['order_total_value']
    .sum()
    .reset_index()
)

customer_revenue.rename(columns={'order_total_value': 'customer_clv'}, inplace=True)

print("Top 5 Customers by CLV:")
customer_revenue.sort_values('customer_clv', ascending=False).head()
```

```python
# =====================================================
# 5.4 – PURCHASE FREQUENCY
# =====================================================

purchase_frequency = (
    orders
    .groupby('customer_id')['order_id']
    .count()
)

print("Average purchase frequency:",
      round(purchase_frequency.mean(), 2))
```

```python
# =====================================================
# 5.5 – AVERAGE BASKET VALUE
# =====================================================

average_basket_value = orders['order_total_value'].mean()

print("Average Basket Value:", round(average_basket_value, 2))
```

```python
# =====================================================
# 5.6 – ITEMS PER ORDER
# =====================================================

items_per_order = (
    order_items
    .groupby('order_id')['order_item_id']
    .count()
)

print("Average items per order:",
      round(items_per_order.mean(), 2))
```

```python
# =====================================================
# 5.7 – REVENUE PER SELLER
# =====================================================

seller_revenue = (
    order_items
    .groupby('seller_id')['total_item_value']
    .sum()
    .reset_index()
)

seller_revenue.rename(columns={'total_item_value': 'seller_total_revenue'}, inplace=True)

print("Top 5 Sellers by Revenue:")
seller_revenue.sort_values('seller_total_revenue', ascending=False).head()
```

```python
# =====================================================
# 5.8 – MONTHLY REVENUE TREND
# =====================================================

orders['purchase_month'] = orders['order_purchase_timestamp'].dt.to_period('M')

monthly_revenue = (
    orders
    .groupby('purchase_month')['order_total_value']
    .sum()
    .reset_index()
)

print("Monthly Revenue:")
monthly_revenue.head()
```

```python
# =====================================================
# 5.9 – MONTHLY ORDER COUNT
# =====================================================

monthly_order_count = (
    orders.groupby('purchase_month')['order_id']
    .count()
    .reset_index()
    .rename(columns={'order_id':'monthly_order_count'})
)

print("Monthly order count:")
monthly_order_count.head()
```
